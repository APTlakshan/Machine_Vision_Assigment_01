```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread('emma_salt_pepper (1).jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# (a
gaussian_filtered = cv2.GaussianBlur(image_rgb, (5, 5), sigmaX=1.0)

# (b) 
median_filtered = cv2.medianBlur(image_rgb, 5)

# Display results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Original noisy image
axes[0].imshow(image_rgb)
axes[0].set_title('Original Image\n(with Salt & Pepper Noise)')
axes[0].axis('off')

# Gaussian filtered image
axes[1].imshow(gaussian_filtered)
axes[1].set_title('(a) Gaussian Smoothing\n(Kernel: 5x5, =1.0)')
axes[1].axis('off')

# Median filtered image
axes[2].imshow(median_filtered)
axes[2].set_title('(b) Median Filtering\n(Kernel: 5x5)')
axes[2].axis('off')



```
