```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def solve_image_tasks(image_path):
    
    img = cv2.imread(image_path)
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    #(a)
    ret, mask = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    print(f"Reported Otsu Threshold Value: {ret}")

   #(b)
    equalized_foreground = gray.copy()
    
    
    foreground_pixels = gray[mask == 255]
    
    
    if len(foreground_pixels) > 0:
        equalized_values = cv2.equalizeHist(foreground_pixels)
        
        equalized_foreground[mask == 255] = equalized_values.flatten()

    # Visualization
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 2)
    plt.title("Otsu Binary Mask")
    plt.imshow(mask, cmap='gray')
    
    plt.subplot(1, 3, 1)
    plt.title("Original Grayscale")
    plt.imshow(gray, cmap='gray')
    
    plt.subplot(1, 3, 3)
    plt.title("Equalized Foreground")
    plt.imshow(equalized_foreground, cmap='gray')
    
    plt.tight_layout()
    plt.show()

    return ret, equalized_foreground

# Execute
threshold_val, final_img = solve_image_tasks('looking_out.jpg')
```
