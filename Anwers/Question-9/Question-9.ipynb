```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image (grayscale)
img = cv2.imread("emma_salt_pepper (1).jpg", cv2.IMREAD_GRAYSCALE)

# Gaussian blur
blur = cv2.GaussianBlur(img, (5, 5), sigmaX=1.0)

# Unsharp masking
k = 1.5  # sharpening strength
sharpened = cv2.addWeighted(img, 1 + k, blur, -k, 0)

# Display results
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Blurred Image")
plt.imshow(blur, cmap='gray')
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Sharpened Image")
plt.imshow(sharpened, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
