```python
import cv2
import numpy as np

def zoom_nearest(img, s):
    h, w = img.shape[:2]
    new_h = int(h * s)
    new_w = int(w * s)

    zoomed = np.zeros((new_h, new_w), dtype=img.dtype)

    for i in range(new_h):
        for j in range(new_w):
            x = int(round(i / s))
            y = int(round(j / s))

            if x >= h:
                x = h - 1
            if y >= w:
                y = w - 1

            zoomed[i, j] = img[x, y]

    return zoomed



img_small = cv2.imread("images_for_zooming/a1q5images/im01.png", cv2.IMREAD_GRAYSCALE)
s = 2.0  

zoom_nn = zoom_nearest(img_small, s)
```

```python
def zoom_bilinear(img, s):
    h, w = img.shape[:2]
    new_h = int(h * s)
    new_w = int(w * s)

    zoomed = np.zeros((new_h, new_w), dtype=np.float32)

    for i in range(new_h):
        for j in range(new_w):
            x = i / s
            y = j / s

            x0 = int(np.floor(x))
            y0 = int(np.floor(y))
            x1 = min(x0 + 1, h - 1)
            y1 = min(y0 + 1, w - 1)

            dx = x - x0
            dy = y - y0

            zoomed[i, j] = (
                (1 - dx) * (1 - dy) * img[x0, y0] +
                dx * (1 - dy) * img[x1, y0] +
                (1 - dx) * dy * img[x0, y1] +
                dx * dy * img[x1, y1]
            )

    return zoomed.astype(img.dtype)
```

```python
img_small = cv2.imread("images_for_zooming/a1q5images/im01.png", cv2.IMREAD_GRAYSCALE)
s = 2.0   # example zoom factor

zoom_nn = zoom_nearest(img_small, s)
```
